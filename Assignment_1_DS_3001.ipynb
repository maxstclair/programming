```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, num=50)
x
```

```python
y = np.log(x)
y
```

```python
z = np.exp(x)
z
```

```python
plt.scatter(x,y,label='Natural Log')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc = 'lower right')
plt.show()
```

```python
plt.scatter(x,z,label='Natural Log')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc = 'lower right')
plt.show()
```

```python
x = np.linspace(-6.5, 6.5, num=100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,label='sin')
plt.plot(x,z,label='cos')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc = 'lower right')
plt.show()
```
